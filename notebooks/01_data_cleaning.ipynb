```python
file_path = "Jupyter Notebook/Hospital Readmission & LOS Analysis/data_raw/diabetic_data.csv"

df = pd.read_csv(file_path)
pd.set_option('display.max_columns', None)
print(df.shape)
print(df.info())

```

```python
df.head()
```

```python
df = df.replace('?', np.nan)

df.isnull().sum()
```

```python
df = df.drop(columns=['weight', 'payer_code', 'medical_specialty'])
print(df.shape)
```

```python
# set reamitted to No and >30 to 0

df['readmitted_binary'] = df['readmitted'].apply(lambda x: 1 if x == '<30' else 0)
print(df['readmitted_binary'].value_counts())
df['readmitted_binary'].value_counts(normalize=True)
```

```python
df['age_clean'] = df['age'].replace({'[0-10)': 5, '[10-20)': 15, '[20-30)': 25, '[30-40)': 35, '[40-50)': 45,
                                     '[50-60)': 55, '[60-70)': 65, '[70-80)': 75, '[80-90)': 85, '[90-100)': 95})
df[['age', 'age_clean']].head(10)
```

```python
# Detail mapping of ICD9 codes to disease categories

def map_icd9_detailed(code):
    try:
        code = float(code)
    except:
        return "Other"
    
    if 390 <= code <= 459 or code == 785:
        return "Circulatory"
    elif 460 <= code <= 519 or code == 786:
        return "Respiratory"
    elif 520 <= code <= 579 or code == 787:
        return "Digestive"
    elif 250 <= code < 251:   # 250.xx
        return "Diabetes"
    elif 800 <= code <= 999:
        return "Injury"
    elif 710 <= code <= 739:
        return "Musculoskeletal"
    elif 580 <= code <= 629 or code == 788:
        return "Genitourinary"
    elif 140 <= code <= 239:
        return "Neoplasms"
    elif 780 <= code <= 799:
        return "Other symptoms"
    elif 240 <= code <= 279 and code != 250:
        return "Endocrine/Metabolic"
    elif 680 <= code <= 709:
        return "Skin"
    elif 1 <= code <= 139:
        return "Infectious"
    elif 290 <= code <= 319:
        return "Mental disorders"
    elif code >= 740 and code <= 759:
        return "Congenital anomalies"
    else:
        return "Other"

for col in ['diag_1', 'diag_2', 'diag_3']:
    df[col] = df[col].apply(map_icd9_detailed)
```

```python
print(df['diag_1'].value_counts(normalize=True).head(10))
print(df['diag_2'].value_counts(normalize=True).head(10))
print(df['diag_3'].value_counts(normalize=True).head(10))
```

```python
df.to_csv("Jupyter Notebook/Hospital Readmission & LOS Analysis/data/diabetic_data_clean.csv", index=False)
```
